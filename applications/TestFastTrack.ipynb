```python
# check checkboxes
def checkCheckboxes():
    for checkbox in checkboxes:
        driver.find_element_by_id(checkbox).click()
    
def secPage():
    driver.find_element_by_class_name("wfPageNextButton").click()
    driver.find_element_by_id("tfa_160").send_keys('TT')
    driver.find_element_by_id("tfa_162").send_keys('TT')
    driver.find_element_by_id("tfa_634").send_keys('TEST')
    driver.find_element_by_id("tfa_636").send_keys('FastTrack')
    driver.find_element_by_id("tfa_637").send_keys('02/17/2000')
    driver.find_element_by_id("tfa_181").send_keys('andy.marks+rega220@macu.edu')
    driver.find_element_by_id("tfa_179").send_keys('8818')
    driver.find_element_by_id("submit_button").click()
    time.sleep(1)
    driver.get("https://midacu.tfaforms.net/217783")
```

```python
# Go through all permutations of bachelors list for traditional students; bl: Bachelors List
def submitTradBachelors(bl):
    for category in bl:
        for k,v in category.items():
            for i in v:  
                for master in masters:                                            
                    checkCheckboxes()
                    driver.find_element_by_id("tfa_15").click()                    
                    driver.find_elements_by_name("tfa_616")[0].click()
                    driver.find_element_by_id("tfa_505").click()            
                    driver.find_element_by_id(k).click()                    
                    driver.find_element_by_id(i).click()
                    driver.find_element_by_id(master).click()
                    secPage()
                    print(k, i, master)

def submitEdu():
    for bach in tradEduB:
        for mas in tradEduM:
            checkCheckboxes()
            driver.find_element_by_id("tfa_15").click()                    
            driver.find_elements_by_name("tfa_616")[1].click()
            driver.find_element_by_id(bach).click()
            driver.find_element_by_id(mas).click()
            secPage()
            print(bach, mas)
            
def submitAdult(location):
    for bach in adultB:
        for mas in masters:
            checkCheckboxes()
            driver.find_element_by_id("tfa_16").click()
            driver.find_element_by_id(location).click()  
            driver.find_element_by_id(bach).click()
            driver.find_element_by_id(mas).click()
            secPage()
            print(bach, mas)
```
